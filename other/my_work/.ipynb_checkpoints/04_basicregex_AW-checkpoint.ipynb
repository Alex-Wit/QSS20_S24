```python
# Read the data from a CSV file into a DataFrame.
schools_df = pd.read_csv("../public_data/schools_df.csv")


# Function to standardize school names based on school type.
def standardize_names(df, school_type, pattern, replacement):
    
    # Filter DataFrame based on a column value (school_type), e.g., 'is_elem_exercise'.
    filtered_df = df[df[school_type] == True].copy()
    
    # Apply a regex pattern to replace terms in the 'schoolname' column and create a new column.
    filtered_df['standardized_name'] = filtered_df['schoolname'].apply(lambda x: re.sub(pattern, replacement, x))
    
    # Return the modified DataFrame.
    return filtered_df



# Define a regex pattern for elementary schools to replace various forms of the word 'elementary'.
elem_pattern = r"(elem.*)(\s+)?(school)?"

# Apply the function to standardize elementary school names.
elem_schools = standardize_names(schools_df, 'is_elem_exercise', elem_pattern, "elemschool")

# Define a regex pattern for charter schools to extract the name part before 'charter'.
charter_pattern = r"(.*)\s+(charter)(\s+)?(\w+)?"

# Apply the function to standardize charter school names, removing 'charter' from the names.
charter_schools = standardize_names(schools_df, 'is_charter_exercise', charter_pattern, r"\1")

# Define a regex pattern for high schools to replace 'high school' or 'hs' with 'highschool'.
hs_pattern = r"(\bhigh\b\s+school|\bhs\b)(\s|$)"

# Apply the function to standardize high school names.
high_schools = standardize_names(schools_df, 'is_highschool_exercise', hs_pattern, "highschool")



# Function to extract the school name that precedes 'highschool'.
def extract_pre_name(df, pattern):
    
    # Apply a regex match to extract everything before 'highschool' and store it in a new column.
    df['pre_highschool_name'] = df['standardized_name'].apply(
        lambda x: re.match(pattern, x).group(1) if re.match(pattern, x) else "")
    
    # Return the DataFrame with the added column.
    return df



# Define a regex pattern to match any characters before 'highschool'.
high_school_pattern = r"^(.*?)\s*highschool$"

# Extract the school name preceding 'highschool' from the standardized names.
high_schools = extract_pre_name(high_schools, high_school_pattern)

# Display the first few rows of the resulting DataFrame to verify results.
print(high_schools[['schoolname', 'standardized_name', 'pre_highschool_name']].head())
```
